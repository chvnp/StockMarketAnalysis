```python
import yfinance as yf
```

```python
msft = yf.Ticker("MSFT")
print(msft)
```

```python
# get stock info
msft.info
```

```python
# get historical market data
msft_data = msft.history(period="max")
```

```python
print(msft_data.shape)
msft_data.head()
```

```python
msft_data.tail()
```

```python
# show actions (dividends, splits)
msft.actions[:10]
```

```python
# show dividends
msft.dividends[:10]
```

```python
# show splits
msft.splits[:10]
```

```python
aapl = yf.Ticker("AAPL")


# get historical market data with dates
aapl_data = aapl.history(start="2019-01-01", end="2019-04-30")
print(aapl_data.shape)
aapl_data.head()
```

```python
# get historical market data interval
# Valid intervals are: 1m, 2m, 5m, 15m, 30m, 60m, 90m, 1h, 1d, 5d, 1wk, 1mo, 3mo
aapl_data1 = aapl.history(interval='60m',start="2019-01-01", end="2019-04-30")
print(aapl_data1.shape)
aapl_data1.head()
```

```python
#download data for multiple tickers at once
mdata = yf.download("SPY AAPL", start="2019-01-01", end="2019-01-10")
```

```python
print(mdata.shape)
mdata.head()
```

```python
#download data for multiple tickers at once with group data by Ticker
gdata = yf.download("SPY AAPL", start="2019-01-01", end="2019-01-10", group_by='ticker')
print(gdata.shape)
gdata.head()
```
